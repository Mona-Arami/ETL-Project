# Extract and Transform
# 32 largest Internet companies from wikipedia


```python
# Use Panda's `read_html` to parse the url
url_largest_Internet_companies = "https://en.wikipedia.org/wiki/List_of_largest_Internet_companies"
table_largest_Internet_companies = pd.read_html(url_largest_Internet_companies, index_col = 0)

Internet_companies_df = table_largest_Internet_companies[1]
Internet_companies_df.drop(["F.Y.","Employees","Headquarters","Founded","Refs"], axis=1, inplace=True)
# Internet_companies_df
```

# 10 largest software companies from wikipedia


```python
url_largest_software_companies = "https://en.wikipedia.org/wiki/List_of_the_largest_software_companies"
table_largest_software_companies = pd.read_html(url_largest_software_companies, index_col = 0)

software_companies_df = table_largest_software_companies[5]
software_companies_df.drop(["Organization","FY","Headquarters"], axis=1, inplace=True)
software_companies_df.rename(
            columns={"Organization.1": "Company", "Sales (B$)": "Revenue ($B)", "Market cap (B$)": "Market Cap. ($B)"}, 
            inplace=True)
# software_companies_df
```

# 16 largest information technology companies from wikipedia


```python
url_largest_IT_companies = "https://en.wikipedia.org/wiki/List_of_the_largest_information_technology_companies"
table_largest_IT_companies = pd.read_html(url_largest_IT_companies, index_col = 0)

IT_companies_df = table_largest_IT_companies[1]
IT_companies_df.drop(["Company","Fiscal Year Ending","Employees","Headquarters","Revenue ($B) USD"], axis=1, inplace=True)
IT_companies_df.rename(columns={"Company.1": "Company"},inplace=True)
IT_companies_df.index.names =["Rank"]
# IT_companies_df
```

```python
# read from ticker.csv file, by merge we add symbol column to datafarames
yahoo_stocks_ticker = pd.read_csv("Resources/ticker.csv", index_col = 0)

Internet_companies_df_merge = pd.merge(yahoo_stocks_ticker, Internet_companies_df, how="right", on="Company")
software_companies_df_merge = pd.merge(yahoo_stocks_ticker, software_companies_df, how="right", on="Company")
IT_companies_df_merge = pd.merge(yahoo_stocks_ticker, IT_companies_df, how="right", on="Company")
# software_companies_df_merge
```

```python
#create list of symboles for each dataframes
largest_Internet_companies_Symbol_list = list(Internet_companies_df_merge["Symbol"])
#delete PRIVATE symbol from list
del largest_Internet_companies_Symbol_list[11]
del largest_Internet_companies_Symbol_list[19]
del largest_Internet_companies_Symbol_list[22]
# largest_Internet_companies_Symbol_list
```

```python
largest_software_companies_Symbol_list = list(software_companies_df_merge["Symbol"])
# largest_software_companies_Symbol_list
```

```python
largest_IT_companies_Symbol_list = list(IT_companies_df_merge["Symbol"])
#delete PRIVATE symbol from list
del largest_IT_companies_Symbol_list[8]
# largest_IT_companies_Symbol_list
```
